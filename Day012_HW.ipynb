### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")
res = {}#用dict來存，key 放分類的類別，value可以以list 的方式放

for d in soup.find(class_="part_list_2").find_all('h3'):
    title = d.find_all('a')[-1].text
    date = d.find(class_="date").text
    tag = d.find_all('em')[-1].text
    res.setdefault(tag,[])#**技巧**使用tag作為key並預設將value值設為[](一開始value沒給值)，以利後之後append {date、append} 進value
    res[tag].append({
        'date' : date ,
        'title': title
                     })
res   
```

### ③ 哪一個類別的文章最多

```python
data = [] #用 list把Dict{分類,數量}包起來
for i in res:
    print(i,len(res[i]))
    d = {} # 用來記錄{分類,數量}
    d['tag'] = i
    d['count'] = len(res[i])
    data.append(d)

for x in sorted(data,key=lambda x:x['count']): #***技巧***sorted與lambda
    print(x)
'''
關於python 與lambda 的使用文章 : https://medium.com/citycoddee/python%E9%80%B2%E9%9A%8E%E6%8A%80%E5%B7%A7-4-lambda-function-%E8%88%87-closure-%E4%B9%8B%E8%AC%8E-7a385a35e1d8
'''    
```
